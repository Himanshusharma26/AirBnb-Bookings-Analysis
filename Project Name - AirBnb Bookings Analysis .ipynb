## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd
airbnb_data = pd.read_csv('Airbnb NYC 2019.csv')
# Data Cleaning
# Drop rows with missing values
airbnb_data.dropna(inplace=True)


#(dropping Rows with Missing Values:The dropna() function is called on the DataFrame airbnb_data.
#The dropna() function drops rows with any missing values (NaN or None) by default.
#The parameter inplace=True is used to modify the DataFrame in place,
 #meaning the changes are applied to airbnb_data directly rather than returning a new DataFrame.

# Remove outliers
#(Removing listings with a price of 0 or minimum nights exceeding 365 enhances dataset quality
# by eliminating potential errors
airbnb_data = airbnb_data[(airbnb_data['price'] > 0) & (airbnb_data['minimum_nights'] <= 365)]

# Convert 'last_review' to datetime format
airbnb_data['last_review'] = pd.to_datetime(airbnb_data['last_review'])
#(Converting 'last_review' to datetime format ensures all dates are in the same style, making them easier to compare.
#It enables simple arithmetic operations with dates, aiding tasks like calculating time differences.
#Datetime format enhances analysis, revealing trends like seasonal patterns and changes over time.
#


# Data Transformation
# Convert 'price' column to numeric
airbnb_data['price'] = airbnb_data['price'].astype(float)

#It ensures that all values in the 'price' column are treated as numbers.
#Numeric data allows for mathematical operations and analysis.
#It helps in maintaining consistency across the dataset.
#Ensures compatibility with numerical operations and calculations.


# Create a new feature 'income' by multiplying price with minimum nights
airbnb_data['income'] = airbnb_data['price'] * airbnb_data['minimum_nights']


# Check if 'room_type' column exists before encoding
if 'room_type' in airbnb_data.columns:
    # Encode categorical variables
    airbnb_data = pd.get_dummies(airbnb_data, columns=['room_type'])


    # Save the cleaned dataset
    airbnb_data.to_csv('cleaned_data.csv', index=False)



else:
    print("Error: 'room_type' column not found in the dataset.")
````

#### Chart - 14 - Correlation Heatmap

A correlation matrix is a table that shows the correlation coefficients between variables in a dataset. Each cell in the table represents the strength and direction of the linear relationship between two variables, with values ranging from -1 to 1. Positive values indicate a positive correlation (both variables increase or decrease together), negative values indicate a negative correlation (one variable increases while the other decreases), and zero indicates no linear correlation. Correlation matrices help identify relationships and patterns between variables, aiding in understanding how different factors influence each other.

correlation_matrix = cleaned_dataset.corr(): This line calculates the correlation matrix for the DataFrame cleaned_dataset. The .corr() method computes pairwise correlation of columns,  and returns a DataFrame representing the correlation matrix.

plt.figure(figsize=(12, 8)): This line creates a new figure for the heatmap plot with a specified size of 12 inches in width and 8 inches in height.

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f"): This line plots the heatmap using seaborn's heatmap() function. It takes the correlation matrix  as input.

annot=True adds annotations (correlation coefficients) to each cell of the heatmap.
cmap='coolwarm' sets the color map to 'coolwarm', which varies from cool colors (e.g., blue) for negative correlation to warm colors (e.g., red) for positive correlation.
fmt=".2f" formats the annotations to display correlation coefficients rounded to two decimal places.

````python

# Calculate correlation matrix
correlation_matrix = cleaned_dataset.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Airbnb NYC 2019 Dataset')
plt.show()
````

##### 1. Why did you pick the specific chart?


The heatmap  was chosen to visualize the correlation matrix because it effectively displays the pairwise correlations between all variables in the dataset. Heatmaps use color gradients to represent the strength and direction of correlations, making it easy to identify patterns and relationships between variables. Additionally, the annotation feature allows for precise interpretation by providing the correlation coefficient values directly on the heatmap. Overall, the heatmap is a powerful tool for exploring and understanding the interdependencies within the dataset.

##### 2. What is/are the insight(s) found from the chart?

Strong Positive Correlation: A correlation coefficient close to 1 indicates a strong positive linear relationship between the variables. In the heatmap, these correlations appear in shades of red.

Strong Negative Correlation: A correlation coefficient close to -1 indicates a strong negative linear relationship between the variables. In the heatmap, these correlations appear in shades of blue.

Weak Correlation: Correlation coefficients close to 0 suggest weak or no linear relationship between the variables. These correlations may appear as white or neutral colors in the heatmap.

Self-Correlation: The diagonal of the correlation matrix represents the correlation of each variable with itself, which is always 1. This is depicted in red to highlight the self-correlation.

Multicollinearity: High correlations between pairs of variables suggest multicollinearity, which can impact the stability and interpretability of regression models.
